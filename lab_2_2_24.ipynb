```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x=np.arange(0, math.pi*2, 0.05)
y1=np.sin(x)
y2=np.cos(x)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0, 0, 1, 1])   # sin x --> graph plotting in graph.  3rd input --> occupy rows.
axes2=fig.add_axes([1, 1, 1, 1])   # cos x --> graph plotting in graph.  4th input --> occupy columns.
axes1.plot(x, y1)                  # [left, bottom, width, height]
axes2.plot(x, y2)
plt.show()
```

```python
ax=plt.axes()
ax.plot([1, 2, 3], label='Line1')   # automatically color aaplied due to legend.
ax.plot([3, 2, 1], label='Line2')
ax.plot([5, 1, 1], label='Line3')
ax.legend()
plt.show()
```
