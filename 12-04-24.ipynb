```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import stats

num_sample = 1000
sample_size = 100

sample_uniform = np.random.uniform(0, 1, (num_sample, sample_size))
means_uniform = np.mean(sample_uniform, axis = 1)

plt.hist (means_uniform, bins = 30, density= True, alpha = 0.5, color = 'b', edgecolor = 'black')

plt.xlabel ('Sample Mean')
plt.ylabel('Probability Density')
plt.title ('Distribution of sample means (uniform distribution)')
plt.show
```

```python
matrix = np.array([[1, 2], [3, 4]])
vector = np.array([1, 1])

result = np.dot (matrix, vector)

plt.figure(figsize = (5, 5))
plt.quiver(0, 0, vector[0], vector[1], angles = 'xy', scale_units= 'xy', scale = 1, color = 'b', label = 'Original')
plt.quiver(0, 0, result[0], result[1], angles = 'xy', scale_units= 'xy', scale = 1, color = 'red', label = 'Result')
plt.xlim(0, result[0]*1)
plt.ylim(0, result[1]*1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Result of Matrix-Vector Multiplication')
plt.grid()
plt.legend()
plt.show()
```
